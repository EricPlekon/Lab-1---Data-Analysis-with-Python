## Loading Data 

```python
import pandas as pd
```

```python
url = 'https://marine.rutgers.edu/~sage/OOI_Data_Workshops/CP02PMUO_R9_ar24011.csv'
data = pd.read_csv(url)
```

```python
# outputs first 5 rows
data.head()
```

```python
# outputs last 5 rows
data.tail()
```

## Extracting columns

```python
# prints out the data from the 'Depth' column
data['Depth'].head()
```

```python
# Extract 2 columns, use double [[]]
data[['Depth', 'Temperature']].head()
```

## Plots

```python
# Plot depth column
data['Depth'].plot()
```

```python
# Plot temperature column
data['Temperature'].plot()
```

```python
# Temperature vs. Depth
ax = data.plot(x='Temperature',y='Depth')
ax.invert_yaxis() # Flips y-axis
```

```python
# Let's go crazy!
data.plot.scatter(x='Salinity', y='Temperature', c='Depth', cmap='viridis_r')
```

```python
# let's use matplotlib directly

import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(data=data, x='Salinity', y='Temperature', c='Depth', cmap='viridis_r')
plt.xlabel('Salinity')
plt.ylabel('Temperature (C)')
cbh = plt.colorbar()
cbh.set_label('Depth (m)')
```

```python
# Histogram chart
data['Temperature'].hist(bins=30)
```

```python
# Boxplots
data[['Temperature', 'Salinity']].plot.box(vert=False)
```

```python
ax = data.plot(x='Oxygen', y='Depth')
ax.invert_yaxis()
```

## Basic Statistics

```python
# .describe() outputs common stats for each variable in dataframe
data.describe()
```

```python
# round these individually
round(data['Temperature'].mean(),4)
```

```python
round(data['Salinity'].std(),3)
```
